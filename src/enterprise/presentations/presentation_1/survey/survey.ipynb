```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Единый контроль размера шрифта
FONTSIZE = 20  # Измените это значение для глобальной настройки

# Конфигурация стиля ДО инициализации Seaborn
plt.rcParams.update({
    'font.size': FONTSIZE,
    'axes.titlesize': FONTSIZE,
    'axes.labelsize': FONTSIZE,
    'xtick.labelsize': FONTSIZE * 0.8,
    'ytick.labelsize': FONTSIZE * 0.8,
    'font.family': 'DejaVu Sans',
    'text.color': 'white',
    'axes.labelcolor': 'white',
    'xtick.color': 'white',
    'ytick.color': 'white',
    'axes.facecolor': 'none',
    'savefig.transparent': True
})

# Инициализация Seaborn с нашими настройками
sns.set_theme(
    style="whitegrid",
    rc={
        'grid.color': '0.4',
        'grid.linestyle': '--'
    }
)

# Создаем папку для графиков
output_dir = "survey_plots"
os.makedirs(output_dir, exist_ok=True)

# Загрузка данных с улучшенной обработкой
def load_data(filepath):
    df = pd.read_csv(filepath, sep=';', skipinitialspace=True, encoding='utf-8')
    return df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

df = load_data('survey2.csv')

def save_single_plot(question, data, filename):
    plt.figure(figsize=(10, 6), facecolor='none')
    ax = plt.gca()
    
    # Обработка данных с учетом множественных ответов
    exploded_data = data.str.split(',\s*').explode()
    perc = exploded_data.value_counts(normalize=True).sort_values(ascending=True) * 100
    
    # Построение горизонтального графика
    plot = sns.barplot(
        x=perc.values,
        y=perc.index.astype(str),
        palette="rocket",
        legend=False,
        dodge=False
    )
    
    # Заголовок графика с формулировкой вопроса
    plt.title(f"{question}", pad=20, fontsize=FONTSIZE, color='white')
    plt.xlabel('Процент ответов', fontsize=FONTSIZE, color='white')
    plt.ylabel('')
    
    # Скрываем рамки и убираем лишние элементы
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.tick_params(axis='y', which='both', length=0, color='white', labelsize=14, labelcolor='white')
    ax.tick_params(axis='x', which='both', labelsize=FONTSIZE, labelcolor='white')
    
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{filename}.png", dpi=300, bbox_inches='tight')
    plt.close()

# Генерация графиков для всех вопросов (исключая столбец "Студент")
for i, column in enumerate(df.columns[1:]):
    save_single_plot(
        question=column,
        data=df[column],
        filename=f"question_{i+1}"
    )

print(f"Графики сохранены в: {output_dir}")
```
